```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression
```

```python
try:
    auto_df = pd.read_csv('Auto (1).csv', na_values='?')
    print("Arquivo 'Auto (1).csv' carregado com sucesso!")
except FileNotFoundError:
    print("Erro: Arquivo 'Auto (1).csv' não encontrado.")
    print("Por favor, certifique-se de que o upload do arquivo para o Google Colab foi concluído e está visível no painel de arquivos.")
    exit()

auto_df = auto_df.dropna()
print(f"O conjunto de dados agora tem {auto_df.shape[0]} linhas após a remoção de valores ausentes.")
print("\nPrimeiras 5 linhas do conjunto de dados:")
print(auto_df.head())
```

```python
print("\n--- Gerando Matriz de Gráfico de Dispersão (Pairplot) ---")
sns.pairplot(auto_df)
plt.suptitle('a. Matriz de Gráfico de Dispersão', y=1.02)
plt.show()
```

```python
print("\n--- b. Matriz de Correlações ---")
# Calcula a correlação de Pearson entre as colunas numéricas.
correlation_matrix = auto_df.corr(numeric_only=True)
print(correlation_matrix)

# Visualização da matriz de correlações com um heatmap para melhor interpretação.
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('b. Matriz de Correlação')
plt.show()
```

```python
print("\n--- c. Análise da Regressão Linear Múltipla ---")
X = auto_df.drop(['mpg', 'name'], axis=1)
y = auto_df['mpg']

model = LinearRegression()
model.fit(X, y)
```

```python
f_statistic, p_values = f_regression(X, y)

print("\ni. Relação entre Preditores e Resposta (Teste F):")
f_test_results = pd.DataFrame({'Preditores': X.columns, 'Estatística F': f_statistic, 'P-valor': p_values})
print(f_test_results)
print("\nComentário: Preditores com p-valor < 0.05 têm uma relação estatisticamente significativa com 'mpg'.")
```

```python
print("\nii. Coeficientes do Modelo (Impacto dos Preditores):")
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coeficiente'])
print(coefficients)
print("\nComentário: Preditores com p-valores baixos (acima) e coeficientes de maior magnitude (positiva ou negativa) são os mais significativos.")
```

```python
year_coefficient = model.coef_[X.columns.get_loc('year')]
print(f"\niii. Interpretação do Coeficiente 'year': {year_coefficient:.4f}")
print(f"Sugestão: Para cada ano que o modelo do carro é mais novo, o consumo ('mpg') aumenta em aproximadamente {year_coefficient:.2f} milhas por galão, mantendo todas as outras variáveis constantes.")
```

```python
print("\n--- d. Gráficos de Diagnóstico ---")
y_pred = model.predict(X)
residuals = y - y_pred
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Valores Ajustados (Previstos)')
plt.ylabel('Resíduos')
plt.title('d. Gráfico de Resíduos vs. Valores Ajustados')
plt.show()
```
